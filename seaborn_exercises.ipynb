### 4. Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.



```python
def get_db_url(database_name):
    from env import host, user, password
    return f'mysql+pymysql://{user}:{password}@{host}/{database_name}'
```

```python
chipotle = pd.read_sql( "SELECT * FROM orders" , get_db_url("chipotle"))
chipotle = chipotle.assign( chipotle_item_price = chipotle.item_price.str.replace("$","").astype("float64") )
```

```python
# The 'most popular' will be defined as the items with the highest quantities sold
chipotle_most_popular = chipotle.groupby("item_name")["quantity"].agg(["sum"]).nlargest(4,"sum")
# Revenue produced by each
chipotle_top_revenue = chipotle.groupby("item_name")["chipotle_item_price"].agg(["sum"]).nlargest(5,"sum")
# Making a mini table based on most popular and revenue
chipotle_popular_revenue= chipotle_most_popular.merge(chipotle_top_revenue, on="item_name")
# Renaming values for clarity
chipotle_popular_revenue = chipotle_popular_revenue.rename(columns={'sum_x': 'total_sold', 'sum_y': 'total_revenue' })

chipotle_popular_revenue
```

```python
sns.catplot(data = chipotle_popular_revenue, x= chipotle_popular_revenue.index, y = "total_revenue", kind="bar")
plt.title("Popular Menu Item Revenue")
plt.xlabel("")
plt.ylabel("Total Revenue (USD)")
plt.xticks(rotation=45);
```
