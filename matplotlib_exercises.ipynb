### 1. Use matplotlib to plot the following equation:

y = x<sup>2</sup> âˆ’ x + 2

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
import matplotlib.pyplot as plt  
import math
```

```python
#Data Points
x = range(-100,100)
y = [x**2 - x + 2 for x in x] # write the list comprehension

#Plot Lines
plt.plot(x,y,
         color='blue', 
         marker='', 
         linestyle= '-' , 
         label = '$y = x^2 - x + 2$')

#Labels
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()

#Table Design
#plt.axvline(0, ls="-", lw=0.5, color = "k")#horizontal line
plt.xticks([-100,-50,0,50,100]);
plt.yticks([y*1000 for y in range(0,10)])

#Annotations 
plt.annotate('(0,0)', 
             xy=(0,0), 
             xytext=(0,1000), 
             arrowprops= dict(facecolor="red"));
```

### 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
#Data Points
x = range(0,10)
y1 = [ math.sqrt(x) for x in x] # write the list comprehension
y2 = [ x**3 for x in x]
y3 = [ 2**x for x in x]
y4 = [1/(x+1) for x in x]

#Figure Details
plt.figure(figsize=(10,10))

#Plot Lines
plt.plot(x,y1,
         color='blue', 
         marker='', 
         linestyle= '-',
         label='y = âˆšx')

plt.plot(x,y2,
         color='red', 
         marker='', 
         linestyle= '-',
         label = 'y = x^3')

plt.plot(x,y3,
         color='green', 
         marker='', 
         linestyle= '-',
         label = 'y=2^x')

plt.plot(x,y4,
         color='orange', 
         marker='', 
         linestyle= '-',
         label = '$y = 1/(x+1)$')

#Labels
plt.title('$y = 1/(x+1)$')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()

#Table Design
plt.axhline(0, ls="-", lw=0.5, color = "k")#horizontal line
plt.axvline(0, ls="-", lw=0.5, color = "k")#vertical line
plt.grid('True', ls = '--')
plt.axis([0, 10, 0, 10])
```
