Curves to graph:

- y = 0n+1      and label the curve "O(1)"
- y = log(n)    and label the curve "O(log n)"
- y = n         and label the curve "O(n)"
- y = nâˆ—log(n)  and label it "O(n log n)"
- y = n<sup>2</sup>        and label it "O(n^2)"
- y = 2<sup>n</sup>        and label it "O(2^n)"
- y = n!        and label it "O(n!)"
- y = n<sup>n</sup> and label it "O(n^n)"


```python
#import pyplot module as alias plt
import matplotlib.pyplot as plt  

#import other libraries 
import numpy as np
import math
```

```python
#DATA
x = range(1,12)
y1 = ([0*n + 1 for n in x])
y2 = ([math.log(n) for n in x])
y3 = ([n for n in x])
y4 = ([n * math.log(n) for n in x])
y5 = ([n**2 for n in x])
y6 = ([2**n for n in x])
y7 = ([math.factorial(n) for n in x])
y8 = ([n**n for n in x])

#Figure Design
plt.figure(figsize=(10,11))

#Plotted Lines
plt.plot(x,y1,
         color='red', 
         marker='', 
         linestyle= '-' , 
         label = 'O$(1)$')
plt.plot(x,y2,
         color='orange', 
         marker='', 
         linestyle= '-' , 
         label = 'O$(log_n)$')
plt.plot(x,y3,
         color='green', 
         marker='', 
         linestyle= '-' , 
         label = 'O$(n)$')
plt.plot(x,y4,
         color='blue', 
         marker='', 
         linestyle= '-' , 
         label = 'O$(n log_n)$')
plt.plot(x,y5,
         color='red', 
         marker='', 
         linestyle= '--' , 
         label = 'O$(n^2)$')
plt.plot(x,y6,
         color='orange', 
         marker='', 
         linestyle= '--' , 
         label = 'O$(2^n)$')
plt.plot(x,y7,
         color='green', 
         marker='', 
         linestyle= '--' , 
         label = 'O$(n!)$')
plt.plot(x,y8,
         color='blue', 
         marker='', 
         linestyle= '--' , 
         label = 'O$(n^n)$')



#Graph Design
plt.title("Big O Notation", fontsize=30)
plt.xlabel("Elements", fontsize=20)
plt.ylabel("Operations", fontsize=20)
plt.legend()# can define a location for legend with plt.legend(loc='upper right', fontsize=14)
plt.axis([1, 10, 0, 20]);
```
